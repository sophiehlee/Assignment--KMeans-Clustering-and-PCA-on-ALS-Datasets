# K-means clustering on the raw data

```python
from sklearn.cluster import KMeans

# Initializing it with KMeans with the number of clusters equal to 3

# Create a k-mean object

kmean = KMeans(n_clusters = 3)
```

```python
# Train model

kmean.fit(data)
```

```python
centroids = kmean.cluster_centers_
print("Shape of Centroids Array: " + str(centroids.shape))

print()
print(centroids)
```

```python
from collections import Counter

labels = kmean.labels_

c = Counter(labels)

for cluster_number in range(0,3):
    print("Cluster {} contains {} samples with percentage of {:2f}%".format(
    cluster_number, c[cluster_number], c[cluster_number]/sum(c.values()) *100))
```

```python
from sklearn.metrics import silhouette_score as ss
```

```python
mean_silhouette_score = ss(data, labels)
print(mean_silhouette_score)
```

```python
# 0.59 is not too bad considering the best value is 1 and the worst value is -1
```

```python
kmean2 = KMeans(n_clusters=2)
kmean2.fit(data)
labels2 = kmean2.labels_
silhouette_score2 = ss(data, labels2)
print(silhouette_score2)
```

```python
kmean7 = KMeans(n_clusters=7)
kmean7.fit(data)
labels7 = kmean7.labels_
silhouette_score7 = ss(data, labels7)
print(silhouette_score7)
```

```python
kmean10 = KMeans(n_clusters=10)
kmean10.fit(data)
labels10 = kmean10.labels_
silhouette_score10 = ss(data, labels10)
print(silhouette_score10)
```

```python
# Cluster of 2 gives a higher silhouette score
```

```python
k_rng = range(1,10)
sse = []
```

```python
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(data)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.plot(k_rng, sse)
```

```python
# k = 2 as per our elbow in the graph
```
